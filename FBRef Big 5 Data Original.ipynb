```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import tkinter as tk
from tkinter import filedialog
```

```python
url = 'https://fbref.com/en/comps/Big5/stats/players/Big-5-European-Leagues-Stats'
df = pd.read_html(url)[0]
#To rename the unnamed columns as ()
for i, columns_old in enumerate(df.columns.levels):
    columns_new = np.where(columns_old.str.contains('Unnamed'), '', columns_old)
    df.rename(columns=dict(zip(columns_old, columns_new)), level=i, inplace=True) 
#To flatten the level 0 and 1
df.columns = [' '.join(col).rstrip('_') for col in [c[::-1] for c in df.columns.values]]

#Allow for full tables to be shown
pd.options.display.max_columns = None
pd.options.display.max_rows = None

#Split the nation names by the space, and use the second one
df['Nation '] = df['Nation '].str.split(' ').str[1]
```

```python
df.head()
```

```python
df.to_csv(r'D:\data analysis study\football\Data collected\updated from python\standar.csv', index = None, header=True)
```
